```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
#Load the Data
df = pd.read_csv('report.csv')
```

```python
#Drop Rows with Missing Data
df_cleaned = df.dropna()  # This will drop all rows with any missing values
```

```python
#Add id Column
df_cleaned.loc[:, 'id'] = range(1, len(df_cleaned) + 1)
```

```python
#Drop ID columnn
df_cleaned = df_cleaned.drop(columns=['ID'])
```

```python
#Array for column order
new_column_order = ['id', 'report_year', 'agency_code', 'agency_jurisdiction', 
                    'population', 'violent_crimes', 'homicides', 'rapes', 
                    'assaults', 'robberies', 'months_reported', 
                    'crimes_percapita', 'homicides_percapita', 
                    'rapes_percapita', 'assaults_percapita', 'robberies_percapita']
```

```python
#Applying new column order
df_cleaned = df_cleaned[new_column_order]
```
